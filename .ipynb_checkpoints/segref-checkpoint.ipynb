```julia
using Pkg
Pkg.activate(".")
Pkg.instantiate()
include("Polyhedron.jl")
using .Poly
using LinearAlgebra, DelimitedFiles, JuMP, HiGHS, NEOSServer 
```

```julia
A = [0.9666 0;
    0.0328 0.9666] 
B = [0.1980;
    0.0033;;]

C = [0 1]

T = 10
ref = 1//20

delta = 0.5

A_exp = hcat(A, [0 0; 0 0])
A_exp = vcat(A_exp, [-T 0 1 T; 0 0 0 delta])

B_exp = vcat(B, [0;0])

E_exp = vcat(zeros(3), [1-delta])

Sx = [1/15 0
      0 1/15;
     -1/15 0;
      0 -1/15;]
Sv = [1/300;
      -1/300;]

Sw = [1/50;
      -1/50;]

Sx = vcat(Sx, zeros(4, 2))
Sv = vcat(vcat(zeros(4, 1), Sv) , zeros(2, 1))
Sw = vcat(zeros(6, 1), Sw)

S = hcat(hcat(Sx, Sv), Sw)

# Fr para a referência
R = [1/15;
      -1/15;]

result = is_pinvariant_seg_ref(A_exp, B_exp, E_exp, S, R; ll = 24)
```

```julia
k = result["K"]

X0 = [1; 1; 0; 0]
r = ones(50)*2'


function trajectory(X0, A, E, r, passos)
    for i in range(2, passos)
        try
            X0 = hcat(X0, (A*X0[i-1] + E*r[i-1]))    
        catch e
            print(e)
        end
        
    end
    return X0   
end

X0 = trajectory(X0, (A_exp + B_exp*k), 50)



```

```julia
include("Polyhedron.jl")
using .Poly

f_plot = [1 0;
        0 1;
        -1 0;
        0 -1]

plot_poly(f_plot, ones(4))

plot!(X0[1, :], X0[2, :],
         seriestype = :path,  
         label="Trajectory",
         linewidth = 2,
        marker = :diamond,
        markersize = 5,
        markercolor = :purple)
```
