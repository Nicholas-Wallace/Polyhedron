```julia
using JuMP, HiGHS
function get_extreme_vertices(A, b, num_pontos=5)
    n_vars = size(A, 2)
    model = Model(HiGHS.Optimizer)
    set_silent(model)
    @variable(model, x[1:n_vars])
    @constraint(model, A * x .<= b)

    lista_vertices = Vector{Vector{Float64}}()

    for i in 1:num_pontos
        c = randn(n_vars)
        @objective(model, Max, sum(c[j] * x[j] for j in 1:n_vars))

        optimize!(model)

        if termination_status(model) == MOI.OPTIMAL
            push!(lista_vertices, value.(x))
        end
    end
    return lista_vertices
end

ext_vertices = get_extreme_vertices(init_cond_F, init_cond_w, )
```

```julia
ext_vertices_tuple = Tuple{Float64, Float64}[]

for i in 1:2:(length(ext_vertices[2]) - 1)
  push!(ext_vertices_tuple, (ext_vertices[2][i], ext_vertices[2][i+1]))
end

display(ext_vertices_tuple)
```
