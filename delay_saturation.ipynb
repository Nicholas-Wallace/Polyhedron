```julia
using Pkg
Pkg.activate(".")
Pkg.instantiate()
include("Polyhedron.jl")
using .Poly
```

```julia
A = [0.9666 0;
    0.0328 0.9666] 
B = [0.1980;
    0.0033;;]

X = [1/15 0;
     0 1/15;
     -1/15 0;
     0 -1/15]

result = Poly.step1_saturation(A, B, X, 20; symetric=false, pond=0.98)

display(result)

#init_cond_F, init_cond_w = Poly.admissable_initCond()
```
